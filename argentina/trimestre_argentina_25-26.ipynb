#### Desglose por compañia de las importaciones de los ultimos 6 meses. Este incluye:

- Nombre de la compañia
- Cantidad de importaciones
- Proveedores, y cantidad de importaciones por proveedor
- Precio promedio de compra por unidad en los ultimos 6 meses
- Volumen total neto de compra en los ultimos 6 meses
- Detalles de la ultima compra realizada


```python
excel_new_data = pd.DataFrame()
```

```python
# FX DEVUELVE DETALLES DE IMPORTADOR PARA el 25

def last_year_data(df):
    global excel_new_data
    companias_mayores_15tm = []

    transition_dic = {
        "NCM/NANDINA/SACH": [],
        "Importador": [],
        "No. Importaciones": [],
        "Volumen Total (Tn)": [],
        "Precio Promedio": [],
        "Ultimo Precio": [],
        # "Proveedores": [],
        "Origenes": [],
        "Ref. Ultima Compra": [],
        "Participacion en Vol. Total": [],
        "Marcas": []
    }

    for company in df['Importador'].unique():

        data = df[df['Importador']
                  == f"{company}"]
        volumenTotalImportacionTn = (
            df['Kgs. Netos'].sum()/1000).round(2)

        data = data.sort_values('Kgs. Netos', ascending=False)

        if (data['Kgs. Netos'].sum() > 15000):

            companias_mayores_15tm.append(company)
            volumenTotalCompania = (data['Kgs. Netos'].sum()/1000).round(2)

            # Adding iterated values to the dataframe
            transition_dic['NCM/NANDINA/SACH'].append(
                ''.join(df['NCM-SIM'].unique())
            )
            transition_dic['Importador'].append(company)
            transition_dic['No. Importaciones'].append(len(data))
            transition_dic['Volumen Total (Tn)'].append(
                volumenTotalCompania
            )
            transition_dic['Precio Promedio'].append(
                (data['U$S Unitario'].mean().round(2))*1000)
            transition_dic['Origenes'].append(
                ', '.join(data['País de Origen'].unique().tolist()))
            transition_dic['Marcas'].append(
                ', '.join(data['Marca - Sufijos'].unique().tolist()))

            data = data.sort_values('Fecha', ascending=False)
            fecha_ultima_compra = data['Fecha'].iloc[0]
            data_ultima_compra = data[data['Fecha'] == fecha_ultima_compra]

            transition_dic['Ultimo Precio'].append(
                (data_ultima_compra['U$S Unitario']._get_value(0, ':,-1'))*1000)
            transition_dic['Ref. Ultima Compra'].append(
                f"Fecha: {fecha_ultima_compra.date()} | Ctdad: {(data_ultima_compra['Kgs. Netos'].sum()/1000).round(2)} tns")
            transition_dic['Participacion en Vol. Total'].append(
                f"{round((volumenTotalCompania / volumenTotalImportacionTn) * 100)}%"
            )

    transition_df = pd.DataFrame.from_records(transition_dic)
    transition_df = transition_df.sort_values(
        by="Volumen Total (Tn)", ascending=False)
    excel_new_data = excel_new_data.append(transition_df)

    print(
        f"\033[1m> Las companias con mas de 15 tm en el ultimo semestre fueron:\033[0m\n\n{companias_mayores_15tm} ")
    print(
        f"\033[1m\n__________________________________________________________________________________\n")
```

```python
# RUNNING MY FX OVER MY 2 DFS
last_year_data(df_new_25)
last_year_data(df_new26_filtered)
last_year_data(df_new_36)
```
